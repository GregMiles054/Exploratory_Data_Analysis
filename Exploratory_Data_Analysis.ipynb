```python
#ADD IMPORT STATEMENTS FIRST
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import the CSV
df = pd.read_csv('dataset.csv') #It is saved in this folder, but if not you would need to give filepath

print(df.shape) #Looking at the rows and columns

df.head(10) #Looking at first 10 rows.
```

```python
#grabbing list of columns
df.columns
#dropping unnamed column using drop function
df = df.drop('Unnamed: 0', axis=1) #axis = 1 means columns
```

```python
df.head()
```

```python
df.dtypes #Checking data types
```

```python
#Checking for null or NA values / Return the same result, is null is an alias for is NA
df.isna() #False means there is data, True means there is null values

df.isna().sum() #Sum of NA cells
```

```python
df.isnull().sum() #Same thing
```

```python
df.loc[df['artists'].isna()]#Looking at an artists columns where the data is NA
```

```python
#Looking at duplicates #Can use keep parameter to keep certain values. setting keep=False means that all duplicate rows, regardless of their position, are marked as duplicates or removed.
df.loc[df.duplicated(keep=False)].sort_values('track_id') #This is returning the ROWS that are duplicated using the LOC function. Notice 894 rows
```

```python
#Looking for duplicates in track id specifically
df.loc[df['track_id'].duplicated(keep=False)].sort_values('track_id')
```

```python
#looking at qualitative varies (Text data, cannot be measured. AKA Catergorical variables)
#Finding unique categories
df['artists'].value_counts()
```

```python
#Looking at quantitative datas (Can be measured, integers and floats)

#Central Tendancy:The location of the middle values (Such as mean, median, mode)
#Spread: How far from the middle the values in a distrubtion fall. Such as Variance, Std Deviation, InterQuartile Range)
#Variance = Mean of the sqwuares of the individual deviations
#Standard Deviation = Square root of the variance
#Quartiles = 3 values that evenly divide a distribution into 4 parts (,25,50,75,10)
#Interquartile range - Q3 minus Q1 <CONSIDERED THE MOST ROBUST MEASURE OF SPREAD, LEAST AFFECTED BY OUTLIERS>
#Skewness: Measure of symmetry in the data
#Kurtosis: Measure of how much of the data falls on the tails or edges of the distribution
```

```python
df.describe()
```

```python
#Plotting danceability distribution via histagram. #Notice the negative skew, meaning more data is to the left.
df['danceability'].hist(bins=60)
```

```python
#VIA box and whisker plot. X needs to be equal to the column that we want (Danceability). #Width of the box is the Inter Quartile range
#Whiskers are the min and max values. 
# Min = Q1 value - 1.5 * Interquartile range. 
# Max = Q3 value + 1.5 * interquartile range


sns.boxplot(x=df['danceability'])
```

Looking at correlation and getting rid of any non numerical data via creating a new data set // AKA MULTIVARIATE ANALYSIS

```python
df.columns #Looking at columns and only selecting numerical data
df_corr = df[[ 'popularity',
       'duration_ms', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature',]].corr()
df_corr
```

Creating a heatmap

```python
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(df_corr, annot=True, cmap='mako') #Providing annotation and changing colors
#Darker value means more negative correlation, lighter value means more positiive correlation (Based on the heatmap index to the right.... purple/peach)
```
